# TRATAMIENTO DE DATOS

Desarrollamos un flujo de trabajo previo que nos garantiza que los datos son consistentes y adecuados para análisis exploratorio y modelos predictivos, así, la preparación de la base de datos realizada tiene principalmente:

## Limpieza y Preprocesamiento de Datos

* Eliminación de caracteres no imprimibles y saltos de línea ocultos en
archivos como postings.csv.
* Manejo de valores nulos y duplicados con estrategias como dropna() o imputación de valores.
* Corrección de errores en nombres de columnas y formatos de datos.
* División del dataset en fragmentos más manejables.

## Normalización y Transformación de Datos

* Conversión de salarios a USD usando tasas de cambio predefinidas.
* Conversión de salarios a una escala anual, considerando períodos de pago (horario, mensual, etc.).
* Normalización y estandarización de variables numéricas.
* Conversión de variables categóricas a valores numéricos con técnicas como one-hot encoding o label encoding.

## Optimización de Tareas

* Creación de funciones auxiliares para automatizar limpieza, transformación y análisis de datos.
* Reducción de dimensionalidad o selección de características para mejorar el rendimiento de modelos.

**Carga y Gestión de Datos en SQLite**

* Extracción y almacenamiento de datos desde archivos CSV/Excel en una base de datos SQLite.
* Creación de tablas estructuradas (postings, companies, salaries, etc.).
* Manejo de valores atípicos y datos inconsistentes antes de su almacenamiento.

## ESQUEMA DE LA BASE DE DATOS

* Atributos de postings.csv:



    job_id: El ID del puesto definido por LinkedIn (https://www.linkedin.com/jobs/view/job\_id)
    company_id: Identificador de la empresa asociada a la oferta de empleo (corresponde a companies.csv)
    title: Título del puesto.
    description: Descripción del puesto.
    max_salary: Salario máximo.
    med_salary: Salario medio
    min_salary: Salario mínimo
    pay_period: Periodo de pago del salario (por hora, mensual, anual)
    formatted_work_type: Tipo de trabajo (Tiempo completo, Tiempo parcial, Contrato)
    location: Lugar de trabajo
    solicitudes: Número de solicitudes que se han presentado
    original_listed_time: Hora original en que se incluyó el puesto en la lista
    remote_allowed: Si el empleo permite el trabajo a distancia
    views: Número de veces que se ha visto la oferta de empleo
    job_posting_url: URL de la oferta de empleo en una plataforma
    application_url: URL a la que se pueden enviar las solicitudes
    application_type: Tipo de proceso de solicitud (externo, complejo/simple in situ)
    expiry: Fecha u hora de caducidad de la oferta de empleo
    closed_time: Hora de cierre de la oferta de empleo
    formatted_experience_level: Nivel de experiencia en el puesto (principiante, asociado, ejecutivo, etc.)
    skills_desc: Descripción de las habilidades requeridas para el puesto
    listed_time: Hora en la que se publicó la oferta de empleo
    posting_domain: Dominio del sitio web con la solicitud
    sponsored: Si la oferta de empleo está patrocinada o promocionada.
    work_type: Tipo de trabajo asociado a la oferta de empleo
    currency: Moneda en la que se proporciona el salario.
    compensation_type: Tipo de retribución del puesto de trabajo.


* Atributos de job_details/benefits.csv

    job_id: El ID del trabajo
    type: Tipo de prestaciones ofrecidas (401K, seguro médico, etc.)
    inferred: Si el beneficio fue explícitamente etiquetado o inferido a través de texto por LinkedIn


* Atributos de company_details/companies.csv


    company_id: El identificador de la empresa definido por LinkedIn
    name: Nombre de la empresa
    description: Descripción de la empresa
    company_size: Agrupación de empresas por número de empleados (0 la más pequeña - 7 la más grande)
    country: País de la sede de la empresa.
    state: Estado de la sede de la empresa.
    city: Ciudad de la sede de la empresa.
    zip_code: Código postal de la sede de la empresa.
    address: Dirección de la sede de la empresa
    url: Enlace a la página de LinkedIn de la empresa

* Atributos de company_details/employee_counts.csv

    company_id: La identificación de la empresa
    employee_count: Número de empleados de la empresa
    follower_count: Número de seguidores de la empresa en LinkedIn
    time_recorded: Hora Unix de recogida de datos
